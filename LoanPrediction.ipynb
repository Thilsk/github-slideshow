**Univariate Analysis**

Examining each variable individually


```python
#Univariate analysis example
train['Loan_Status'].value_counts().plot.bar()
```

```python
#Visualizing Categorical Independent Variables
plt.figure(1)
plt.subplot(221)
train['Gender'].value_counts(normalize=True).plot.bar(figsize = (20,10),title = 'Gender')
plt.subplot(222)
train['Married'].value_counts(normalize=True).plot.bar(figsize = (20,10),title = 'Married')
plt.subplot(223)
train['Self_Employed'].value_counts(normalize=True).plot.bar(figsize = (20,10),title = 'Self_Employed')
plt.subplot(224)
train['Credit_History'].value_counts(normalize=True).plot.bar(figsize = (20,10),title = 'Credit_History')
plt.show()

```

```python
#Visualizing Ordinal Independent Variables
plt.figure(1)
plt.subplot(131)
train['Dependents'].value_counts(normalize=True).plot.bar(figsize = (24,6),title = 'Dependents')
plt.subplot(132)
train['Education'].value_counts(normalize=True).plot.bar(figsize = (24,6),title = 'Education')
plt.subplot(133)
train['Property_Area'].value_counts(normalize=True).plot.bar(figsize = (24,6),title = 'Property_Area')
plt.show()
```

```python
#Visualizing Numerical Independent Variables
#Applicant Income
plt.figure(1)
plt.subplot(121)
sns.distplot(train['ApplicantIncome'])
plt.subplot(122)
train['ApplicantIncome'].plot.box(figsize = (16,5))
plt.show()
```

```python
#Segragating Applicant Income by Education
train.boxplot(column = 'ApplicantIncome',by = 'Education')
plt.suptitle("")
```

```python
#Coapplicant Income
plt.figure(1)
plt.subplot(121)
sns.distplot(train['CoapplicantIncome'])
plt.subplot(122)
train['CoapplicantIncome'].plot.box(figsize = (16,5))
plt.show()
```

```python
#Loan Amount
plt.figure()
plt.subplot(121)
sns.distplot(train['LoanAmount'])
plt.subplot(122)
train['LoanAmount'].plot.box(figsize = (16,5))
```
